 * 결측치는 데이터에 NaN값으로 존재할 수 도 있지만, 때로는 의미없는 숫자들(0 or inf 등 column의 의미와 맞지 않은)로 차있는 경우도 있습니다.
 * 데이터 탐색을 통해 pima dataset의 결측치를 탐색해 봅니다.

### 1-1) (자유방식, df1)결측치 혹은 이상치를 갖는 row를 유추해주세요<br>(Hint : 각 Columns들의 기초 통계값을 관찰해봅시다.mean,std,min,max,25,50,75% quantile)  <br> 1-2) Column1~5이 0의 값을 갖는 데이터들이 결측치로 가정해보고, 이 데이터들을 실제 결측치로 표시(np.nan)해서 df2에 저장해주세요 <br> 1-3) df2에서 결측치가 포함된 row들을 제거해df3에 저장해 주세요 <br> 1-4) df3의 통계값들을 관찰해본 후 df1과 비교해보세요(주석에 의견을 적어주세요)<br>1-5) 결측치를 다루는 또 다른 방식은 Imputation이 있습니다. df5에 df2의 결측치들을 평균값으로 대체해 주세요


```python
df1 = pd.read_csv('Pima_dataset.csv')

################################################################################
# 1) Code :                                                                    # 
################################################################################
#1-1
print(df1.describe())
# 해석 : count는 모두 768개로 동일하지만, Column1~5는 mean과 std, quantile(25,50,75)를 살펴봤을 때 
#       0의 값은 현저히 주 분포에서 떨어진 수치들 입니다.(Column1~5의 min=0이기 때문에 0이 존재한다는것을 알 수 있음)
#       또한 각 column의 의미를 봤을 때 0의 값이 올 수 없는 column들도 있음(ex. column5 몸무게)
print((df1.iloc[:,[1,2,3,4,5]] == 0).sum()) # Column1~5에서 결측치가 의심되는 Count
# 해석 : count결과 column3,4 는 데이터의 1/3~1/2 정도의 수치로 결측치가 아닐 가능성이 있습니다.
#       이와 같은 경우, column들의 의미를 살펴봐야 합니다.

#1-2
df2 = df1.copy()
df2.iloc[:,[1,2,3,4,5]] = df2.iloc[:,[1,2,3,4,5]].replace(0,np.nan)
display(df2)

#1-3
df3 = df2.dropna(axis=0) 
df3 = df3.reset_index() # index 초기화
del df3['index']
display(df3)

#1-4
print(df3.describe())

#1-5
df5 = df2.copy()
df5.fillna(df5.mean(), inplace=True)
display(df5)
################################################################################
#                                 END OF YOUR CODE                             #
################################################################################
```

## Outlier Practice (2~3)
 * Outlier는 Missing Value와 함께 데이터 탐색 단계에서 꼭 고려해야 합니다.
 * Outlier를 탐색을 통계값관찰, 시각화, boxplot 의 측면에서 실습하는 과제입니다.
 * 뒷주차에 배우는 clustering 방법론을 활용한 outlier detection도 효율적인 방법론이지만, 아직 고려하지 않습니다.

### 하나의 변수에 대한 Outlier 탐색 <br><br>2-1) X의 mean, std를 살펴본 후 mean+-3std 밖에 벗어난 이상치를 구해주세요<br> 2-2) plt.hist()함수를 통해서 이상치를 유추해주세요(주석에 의견을 적어주세요) <br> 2-3) sns.boxplot() 함수를 사용해서 이상치를 유추해 주세요(주석에 의견을 적어주세요)

```python
X = np.load('./data/outlier1.npy')
################################################################################
# 2) Code :                                                                    # 
################################################################################
#2-1
print('평균 : ' , np.mean(X))
print('표준편차 : ' , np.std(X))
# 정규분포를 가정했을 때, 평균에서 +- 3표준편차 내의 데이터들은 약 99%이상이 포함되는 구간입니다.
# 현실의 데이터가 정확한 정규분포는 아니지만, 일반적으로 이런 데이터들을 outlier로 의심해 볼 수 있습니다.
print('평균+3표준편차 보다 큰 데이터 : ', X[np.where(X > np.mean(X) + 3 * np.std(X))])
print('평균-3표준편차 보다 작은 데이터 : ', X[np.where(X < np.mean(X) - 3 * np.std(X))])

#2-2
fig1=plt.figure()
plt.hist(X)
# 해석 : 시각화(histogram)결과 500과 800 부근에 동떨어진 값들이 관찰되는 것을 볼 수 있습니다.

#2-3
fig2=plt.figure()
sns.boxplot(X)
# 해석 : 시각화(boxplot)의 결과 -400, 500, 800 부근에 돌떨어진 값들이 관찰되는 것을 볼 수 있습니다.
#       Q1, Q3로 부터 IQR의 1.5배 밖에 있는 데이터를 outlier로 볼 수 있기 때문에,
#       boxplot결과 해석으로 이 3가지 데이터포인트를 outlier라 의심해볼 수 있다.

# 결과적으로 원하는 기준에 따라서 outlier는 조금씩 달라질 수 있고, 데이터 탐색 단계에서 반드시 고려한 후 넘어가야 합니다.
################################################################################
#                                 END OF YOUR CODE                             #
################################################################################
```
